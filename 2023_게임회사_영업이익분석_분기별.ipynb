```python
df_4.columns = ['매출액(4분기)', '영업이익(4분기)', '당기순이익(4분기)']
df_3.columns = ['매출액(3분기)', '영업이익(3분기)', '당기순이익(3분기)']
df_2.columns = ['매출액(2분기)', '영업이익(2분기)', '당기순이익(2분기)']
df_1.columns = ['매출액(1분기)', '영업이익(1분기)', '당기순이익(1분기)']
```

```python
print(df_4)
```

# 회사별 2023년도 영업이익 분기별 비교

```python
df_영업이익비교 = pd.concat([df_1['영업이익(1분기)'], df_2['영업이익(2분기)'], df_3['영업이익(3분기)'], df_4['영업이익(4분기)']], axis = 1)
```

```python
print(df_영업이익비교)
```

#### 

```python
df_영업이익비교.index
```

```python
print(type(df_영업이익비교.index))
```

```python
new_index = []

for i in df_영업이익비교.index:
    j = i[::-1]
    k = j[8::]
    l = k[::-1]
    new_index.append(l)
    
print(new_index)
```

```python
df_영업이익비교.index = new_index
```

```python
print(df_영업이익비교.index)
```

```python
from matplotlib import rc
import platform

if platform.system() == 'Windows':
    rc('font', family='Malgun Gothic')
elif platform.system() == 'Darwin': # Mac
    rc('font', family='AppleGothic')
else: #linux
    rc('font', family='NanumGothic')
    
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 현상 해결 코드
```

```python
df_st = df_영업이익비교.sort_values(by = ['영업이익(4분기)'], ascending = True)

df_st.plot(title = '2023년 분기별 영업이익 비교', xlabel = '회사명', kind = 'barh', xticks = np.arange(-100000000000,900000000000,100000000000), width = 0.65, figsize=(8, 10))
```
